***1.Heart disease percentage plot***

```python
HC=df['HeartDisease'].value_counts()
HC
```

```python
plt.figure(figsize=(5,5))
wedgeprops={'edgecolor':'red','linewidth':3}
plt.pie(HC,labels=HC.index,wedgeprops=wedgeprops,colors=['black','white'],explode=(0,0.1),autopct="%1.2f%%",shadow=True)
plt.legend(['Yes','No'],loc='lower right')
plt.title('HEART DISEASE COUNT')
plt.show()

```

INSIGHT:In these pi-Chart we have to see that most of the population have does not heart disease.

***2.Smocking count plot***

```python
SC=df['Smoking'].value_counts()
SC
```

```python
plt.figure(figsize=(5,5))
sns.countplot(x=df['Smoking'])
plt.title('count plot of smocking',color='red')
plt.show()
```

In these plot we have to see that count of smocking population, most of population are not smocking .

 ***3.Stroke level of alcohol drinking and smoking peoples***

```python
plt.figure(figsize=(5,5))
sns.countplot(x=df['Smoking'],data=df,hue=df['Stroke'])
plt.show()
```

stroke people with smokingis comparitievely grater than stroke people without smoking

***5.High risk people by gender***

```python
HR=df[(df["Smoking"]=="Yes") & (df["HeartDisease"]=="Yes") & (df["AlcoholDrinking"]=="Yes") & (df["Diabetic"]=="Yes") & (df["PhysicalActivity"]=="No") & (df["BMI"]>30)]
HRS=HR["Sex"].value_counts()
HRS
```

```python
plt.figure(figsize=(5,5))
plt.pie(HRS,labels=HRS.index,colors={'violet','pink'},explode=(0,0),autopct='%1.2f%%')
plt.title('Gender diffrence in high risk people')
plt.show()
```

In these pie graph shows males health are high risk.

***6.Age distribution of smoking***

```python
SM=df[(df["Smoking"]=="Yes") & (df["HeartDisease"]=="Yes") & (df["AlcoholDrinking"]=="Yes")]
```

```python
SMC=SM['AgeCategory'].value_counts()
SMC
```

```python
plt.style.use('dark_background')
plt.figure(figsize=(10,10))
plt.title('Age distribution of smoking')
plt.bar(SMC.index,SMC.values,color='cyan')
plt.xticks(rotation=45)
plt.show()
```

The number of aged peoples with Heart disease who are Alcholics and smokers are greater

***7.DISTRIBUTION PLOT OF MENTAL HEALTH AMONG HIGH RISK HEART PATIENTS***


```python
HR=df[(df["Smoking"]=="Yes") & (df["HeartDisease"]=="Yes") & (df["AlcoholDrinking"]=="Yes") & (df["Diabetic"]=="Yes") & (df["PhysicalActivity"]=="No") & (df["BMI"]>30)]
HR
```

```python
plt.figure(figsize=(10,10))
sns.displot(HR['MentalHealth'],kde=True,bins=3,color='violet')
plt.title('Distribution plot',color='g')
plt.show()
```

INSIGHT-The majority of heart patients fall within the 0-10 range on the Mental health scale.

***8.Distribution plot of General health among High risk heart patients.***

***9.Comparison of General Health Condition by agecatagory.***

```python
Healthy=df[(df["Smoking"]=="No") & (df["HeartDisease"]=="No") & (df["AlcoholDrinking"]=="No") & (df["SkinCancer"]=="No") & (df["Stroke"]=="No") & (df["Diabetic"]=="No") & (df["PhysicalActivity"]=="Yes") & (df["BMI"]<24) & (df["KidneyDisease"]=="No")]
GH=Healthy['GenHealth'].value_counts()
```

```python
plt.figure(figsize=(10,10))
sns.countplot(x='AgeCategory',data=Healthy,hue='GenHealth')
plt.title('Healthy individuals by Age catagory')
plt.tight_layout()
```

INSIGHT- Excellent health of people between 18-24 age catagory.
